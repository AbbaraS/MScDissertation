```python
import os
import pandas as pd

normal_data_path = "../data/Inputs/normal_cases/"
metadata_list = []

for folder_name in sorted(os.listdir(normal_data_path)):
    if "_" in folder_name and folder_name != ".DS_Store":  # Check if name follows format (PatientID_AgeGender)
        patient_id, age_gender = folder_name.rsplit("_", 1)
        print(folder_name)
        print(age_gender)
        age = int("".join(filter(str.isdigit, age_gender)))  # Extract numbers (age)
        gender = 1 if "F" in age_gender else 0  # Encode Female=1, Male=0

        metadata_list.append([patient_id, age, gender])

# Save to CSV
metadata_df = pd.DataFrame(metadata_list, columns=["PatientID", "Age", "Gender"])
metadata_df.to_csv("normal_cases_metadata.csv", index=False)

print("✅ Normal cases metadata saved to CSV!")
```

```python
import pandas as pd

# Load the CSV file while removing empty columns
file_path = "takotsubo_cases_metadata_dirty.csv"

# Read CSV and drop completely empty columns
df = pd.read_csv(file_path).dropna(axis=1, how="all")

# Keep only relevant columns: Initials, MRN (Patient ID), Age, Sex
df = df[["Initials", "MRN", "Age", "Sex"]]

# Combine Initials and Patient ID into one column
df["PatientID"] = df["Initials"] + " " + df["MRN"].astype(str)

# Convert Gender to numerical values (F → 1, M → 0)
df["Gender"] = df["Sex"].map({"F": 1, "M": 0})

# Reorder columns so that PatientID is first
df = df[["PatientID", "Age", "Gender"]]

# Save the cleaned version
cleaned_file_path = "takotsubo_cases_metadata.csv"
df.to_csv(cleaned_file_path, index=False)

print(f"✅ Cleaned metadata saved as {cleaned_file_path}!")
```
